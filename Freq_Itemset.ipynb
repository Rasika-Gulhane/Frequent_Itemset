## Generating Frequent itemset

```python
#importing general packages
import numpy as np
import pandas as pd
```

```python
#
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
additional_data=[['Coffee','Oil','Eggs','Potatoes','Corn'],
                ['Onion','Nutmeg','Coffee','Milk'],
                ['Banana','Oil', 'Coffee','Eggs','Corn'],
                ['Dill','Apple','Onion','Nutmeg','Butter'],
                ['Kidney Beans','Butter','Eggs','Nutmeg','Ice cream','Onion']]
```

```python
New_dataset= dataset+additional_data
```

```python
%pip install mlxtend --upgrade
# restart kernel after this
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
# The fit method is to learn what is the item to transaction
# the transform method is to Transform transactions into a one-hot encoded NumPy array.
te_ary = te.fit(New_dataset).transform(New_dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.5 , use_colnames=True)
```

## Selecting and Filtering Result

From this, it is depicted that Eggs has highest occurance with 70%,  
and Nutmeg has 50% occurance while Onion and kidney Beans each has 60%. 
we can see Eggs with KidneyBeans has 50% buying rate 
also nutmeg with onion also has 50% buying rate

```python
frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.3) ]
```

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == {'Onion', 'Eggs'} ]
```

## Association Rule Generation from Selected Item

```python
from mlxtend.frequent_patterns import association_rules

associationRulesSet =association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
associationRulesSet
```

```python
for i , row in associationRulesSet.iterrows():
    print(f"{i+1} - {list(row.antecedents)} -> {list(row.consequents)} = [support = {row.support} , confidence = {row.confidence}, lift ={row.lift}] , dependent = {row.lift != 1.0}" )
```

```python
associationRulesSet["antecedent_len"] = associationRulesSet["antecedents"].apply(lambda x: len(x))
associationRulesSet
```

```python
associationRulesSet[associationRulesSet['antecedents'] == {'Kidney Beans'}]
```

```python
associationRulesSet[associationRulesSet['consequents'] == {'Eggs'}]
```

```python
associationRulesSet[(associationRulesSet['support'] >= 0.5) &
                    (associationRulesSet['confidence'] >=0.8)&
                    (associationRulesSet['antecedent_len'] ==1) ]
```

As we consider check with query by rule :

- rules 
    - support >= 0.5
    - confidence >= 0.8
    - lenth of items as antecedents = 1

## Working with Sparse Regression

```python
# To save memory, you may want to represent your transaction data in the sparse format. This is especially useful if you have lots of products and small transactions.

oht_ary = te.fit(dataset).transform(dataset, sparse=True)
sparse_df = pd.DataFrame.sparse.from_spmatrix(oht_ary, columns=te.columns_)
sparse_df
```

```python
apriori(sparse_df, min_support=0.6, use_colnames=True, verbose=1)
```

## frequent Itemset via FP growth algorithm
